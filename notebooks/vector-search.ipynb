```python
raw_documents = TextLoader("tagged_description.txt", encoding="utf-8").load()
text_splitter = CharacterTextSplitter(chunk_size=0, chunk_overlap=0, separator="\n")
documents = text_splitter.split_documents(raw_documents)
```

```python
documents[0]
```

```python
import os
from dotenv import load_dotenv
load_dotenv()
embedding = GoogleGenerativeAIEmbeddings(
    model="models/embedding-001", 
    google_api_key=os.getenv("GOOGLE_API_KEY")  # assuming you loaded it from .env
)
db_books = Chroma.from_documents(documents, embedding=embedding, persist_directory="./chroma_db")
```

```python
query = "A book to teach children about nature"
docs = db_books.similarity_search(query, k = 10)
docs
```

```python
books[books["isbn13"] == int(docs[0].page_content.split()[0].strip())]
```

```python
def retrieve_semantic_recommendations(
        query: str,
        top_k: int = 10,
) -> pd.DataFrame:
    recs = db_books.similarity_search(query, k = 50)

    books_list = []

    for i in range(0, len(recs)):
        books_list += [int(recs[i].page_content.strip('"').split()[0])]

    return books[books["isbn13"].isin(books_list)]
```

```python
retrieve_semantic_recommendations("A book to teach children about nature")
```

```python
from langchain.vectorstores import Chroma
from langchain_google_genai import GoogleGenerativeAIEmbeddings
from langchain.text_splitter import CharacterTextSplitter
from langchain.docstore.document import Document
import pandas as pd
import os

# Load your books CSV with the text to embed, e.g. use 'tagged_description' or any text field
books = pd.read_csv(r"C:\Users\anupa\Desktop\BR\backend\books\books_cleaned.csv")

# Prepare the documents: split large text if needed
texts = books['tagged_description'].tolist()  # or any text column you want to embed
isbn13s = books['isbn13'].tolist()

documents = []
for isbn, text in zip(isbn13s, texts):
    # Prepend ISBN13 to text to keep track of it in the vectorstore metadata
    full_text = f"{isbn} {text}"
    documents.append(Document(page_content=full_text))

# Initialize embedding model
embedding = GoogleGenerativeAIEmbeddings(
    model="models/embedding-001",
    google_api_key=os.getenv("GOOGLE_API_KEY")
)

# Create vector store and add documents
db = Chroma.from_documents(documents, embedding=embedding, persist_directory="../notebooks/chroma_db")

# Persist the DB to disk
db.persist()

print(f"DB contains {db._collection.count()} vectors")
```
